### Задание 6. 
Предположим, нулевой гипотезой вашего исследования являлось предположение, что конверсия в генеральной совокупности равняется 0.4. Вы получили p = 0,12 и не смогли отклонить нулевую гипотезу. Однако позже выяснилось, что конверсия в генеральной совокупности действиетльно равна 0.4. Как можно оценить результаты?

Какой вариант корректный по вашему мнению и почему?

**V** 1) Вы не совершали ни ошибку первого рода, ни ошибку второго рода.

2) Вы совершили ошибку первого рода

3) Вы совершили ошибку второго рода


#### Ответ

Первое утверждение верное.

* Ошибка первого рода - отклонили нулевую гипотезу, когда она была верна. 
* Ошибка второго рода - не отклонили нулевую гипотезу, хотя была верна альтернативная.


1. У нас есть нулевая гипотеза, в рамках которой мы совершаем исследование. Наша выборка обладает р-уровнем значимости равным 0,12.
2. р = 0,12 не дает нам права отказаться от нулевой гипотезы и принять альтернативную (какой бы она не была). Результатом нашего исследования является принятие, а не отклонение нулевой гипотезы.
3. Уже постфактум мы узнаем, что наша нулевая гипотеза оказалась верной. Другими словами, р-уровень нашей выборки заставил нас придерживаться верной нулевой гипотезы, т.е. мы не отказались от верной гипотезы. Мы не совершали ошибок.
